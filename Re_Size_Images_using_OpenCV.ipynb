### 2. Resize the 100 images to 50%
---

```python
import cv2
import os

# Define the directory path where the images are located
directory = "/content/dataset/Bird"

# Loop through all the files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Read the image
        img_path = os.path.join(directory, filename)
        img = cv2.imread(img_path)

        # Get the original image dimensions
        height, width = img.shape[:2]

        # Calculate the new dimensions for resizing
        new_width = int(width * 0.5)
        new_height = int(height * 0.5)

        # Resize the image
        resized_img = cv2.resize(img, (new_width, new_height))

        # Save the resized image with a new filename
        new_filename = "resized_" + filename
        new_img_path = os.path.join(directory, new_filename)
        cv2.imwrite(new_img_path, resized_img)

        print(f"Resized {filename} to 50%: {new_filename}")
```
